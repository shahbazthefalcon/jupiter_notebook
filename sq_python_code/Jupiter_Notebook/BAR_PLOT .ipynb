# HOW to Set the order Male Female etc [order0["Feamale,male"]]

```python
import seaborn as sns
import matplotlib.pyplot as pl
big_ship=sns.load_dataset("titanic")
big_ship

sns.barplot(x="sex",y="alone", data=big_ship,hue="who", order=["female","male"])
pl.show
```

# ERROR BARS REMOVAL (Confidence interval ci=none)

```python
import seaborn as sns
import matplotlib.pyplot as pl
big_ship=sns.load_dataset("titanic")
big_ship

sns.barplot(x="sex",y="alone", data=big_ship,hue="who", order=["female","male"],color="red",ci=None)
pl.show
```

```python
# ESTIMATOR MEAN MEDIAN MODE IS FUNCTION OF NUMPY SO IMPORT NUMPY
```

```python
import seaborn as sns
import numpy
import matplotlib.pyplot as pl
big_ship=sns.load_dataset("titanic")
big_ship

sns.barplot(x="class",y="fare", hue="sex", data=big_ship,estimator=median)
pl.show
```

# SATURATION AND INTENSITY OF COLORS

```python
import seaborn as sns
import numpy
import matplotlib.pyplot as pl
big_ship=sns.load_dataset("titanic")
big_ship

sns.barplot(x="class",y="fare", hue="sex", data=big_ship,estimator=median, saturation=1)
pl.show
```
