# ONLY USE THE FOLLOWING CODES THE REST WILL BE RUN FROME ABOVE CODE SHOWS THE MEANS)

```python
# import seaborn as sns
# import matplotlib.pyplot as pl
# import pandas as pd
# import numpy as np
big_ship=sns.load_dataset("titanic")
sns.boxplot(x= "survived",y= "age", showmeans=True,
            meanprops={"marker":"+",
                        "markersize":"12",
                        "markeredgecolor":"white"}, 
            data=big_ship)
pl.xlabel("How many survived",size=14)
pl.ylabel("Age(years)",size=12)
pl.title("BOX PLOT of TITANIC SURVIVED",size=14,weight="bold")
pl.show()
```
