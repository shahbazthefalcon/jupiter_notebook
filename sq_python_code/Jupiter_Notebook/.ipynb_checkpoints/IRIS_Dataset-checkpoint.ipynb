# IMPORT LIBRARIES

```python
import seaborn as sns
import matplotlib.pyplot as pl
```

**IMPORTINg DATA SET IRIS FLOWERS DATASET**

```python
flowers_data=sns.load_dataset("iris")
flowers_data
```

```python
#DRAW LINE PLOT
```

```python
sns.lineplot(x="sepal_length",y="sepal_width", data=flowers_data)
pl.show
```

```python
# ADDING TITLE
```

```python
sns.lineplot(x="sepal_length",y="sepal_width", data=flowers_data)
pl.title("Chart of Flowers")
pl.show
```

```python
## ADDING LIMITS OF X AND Y
```

```python
sns.lineplot(x="sepal_length",y="sepal_width", data=flowers_data)
pl.xlim(2)
pl.ylim(1)
pl.show
```

```python
## SET STYLE
1.darkgrid
2.whitegrid
3.dark4
4.white
5.ticks
```

```python
sns.lineplot(x="sepal_length",y="sepal_width", data=flowers_data)
sns.set_style("darkgrid")
sns.set_style(style=None,rc=None)  #This will change style to dark, Also try other styles 
pl.show
```

```python
# SET THE PLOT SIZE
```

```python
flowers_data=sns.load_dataset("iris")
pl.figure(figsize=(10,6))#This describes the size of the figure after the dataset is defined 
sns.lineplot(x="sepal_length",y="sepal_width", data=flowers_data)
sns.set
pl.show
```
